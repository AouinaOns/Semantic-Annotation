```python

nlp = spacy.load('fr_dep_news_trf')
```

```python
def create_sentence_list(text,nlp):
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]
    return sentences

def split_into_paragraphs(text):
    paragraphs = text.split('\n\n')  # Split based on double line breaks
    return paragraphs


def read_text_files_in_directory(directory):
    """
    Reads text files in a directory and returns their contents along with their filenames.

    Args:
        directory (str): The directory containing the text files.

    Returns:
        list: A list of tuples, each containing the filename and the contents of a text file.
    """
    text_files = []

    # Loop through the files in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):  # Check if the file is a text file
            file_path = os.path.join(directory, filename)
            file_name_without_extension, file_extension = os.path.splitext(filename)

            with open(file_path, "r", encoding="utf-8") as file:
                text = file.read()
                text_files.append((file_name_without_extension, text))

    return text_files


import re


def find_chunks_with_regex(text, chunk):
    try:
        # Remove spaces around hyphens between words
        chunk = re.sub(r'\s*-\s*', '-', chunk)

        # Remove extra space after an apostrophe
        chunk = re.sub(r"(l|d)('|’)\s+", "", chunk)
        
        # Remove extra space after an apostrophe
        #chunk = re.sub(r"'\s+", "’", chunk)

        
        pattern = re.compile(chunk, re.IGNORECASE)
    except re.error as e:
        print("Invalid regex pattern:", e)
        return []

    matched_spans = []
    
    for match in pattern.finditer(text.lower()):
        start, end = match.span()
        matched_text = text[start:end]
        matched_spans.append((matched_text, start, end))
    
    if not matched_spans:
        print("No matches found for:", chunk)
    
    return matched_spans


    return matched_spans


def convert_to_brat_ann(spans):
    lines = []
    for idx, (span, start, end) in enumerate(spans, start=1):
        #entity_type = span.entity_type
        start = start
        end = end # End position is exclusive in spaCy
        text = span
        line = f'T{idx}\t{"annotationToAdd"} {start} {end}\t{text}\n'
        lines.append(line)
    return ''.join(lines)

def write_brat_file(text, output_directory, file_name, spans):
    """
    Write text to a BRAT .ann file in the specified output directory.

    Args:
        text (str): The text content to write to the BRAT file.
        output_directory (str): The directory where the BRAT file will be saved.
        file_name (str): The name of the BRAT file (without the extension).
        spans (list): List of spans to convert to BRAT annotation format.
    """
    brat_ann = convert_to_brat_ann(spans)
    # Create the BRAT .ann file path
    brat_file_path = os.path.join(output_directory, f"{file_name}.ann")
    # Write the BRAT annotation content to the .ann file
    with open(brat_file_path, 'w+', encoding='utf-8') as ann_file:
        ann_file.write(brat_ann)
```
